```python
import numpy as np
import matplotlib.pyplot as plt

V_charging=np.array([0, 3.82, 6.02,7.75,8.87,9.61,10.17,10.52,10.76,10.93,11.03,11.11,11.16])
t=np.arange(0,125,10)

V_discharging=np.array ([12.02,8.63,5.84,4.02, 2.74,1.9,1.34,0.92,0.66,0.47,0.33,0.23,0.17])
```

```python
plt.xlabel("Time in Seconds")
plt.ylabel("Voltage in Volts")
plt.title("Charging")
plt.plot(t,V_charging)

```

```python
plt.xlabel("Time in Seconds")
plt.ylabel("Voltage in Volts")
plt.title("Discharging")
plt.plot(t,V_discharging)



```
